```python
df['age'].fillna(df['age'].mean(), inplace=True)
df
```

```python
df.dtypes
```

```python
df = df.astype({
    'survived': 'int64',
    #'sex': 'string',
    'age': 'float64',
    'sibsp': 'int64',
    'parch': 'int64',
    'fare': 'float64',
    'embarked': 'string'
})
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['age'].hist(bins=20)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Age')
plt.show()
```

```python
sns.barplot(x='class', y='survived', data=df)
plt.xlabel('Pclass')
plt.ylabel('Survived')
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
sns.catplot(x='class', y='survived', hue='sex', kind='bar', data=df)
plt.xlabel('Pclass')
plt.ylabel('Survived')
plt.title('Survival Rate by Passenger Class and Sex')
plt.show()
```

```python
sns.countplot(x='survived', data=df)
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Count of Survived Passengers')
plt.show()
```

```python
# Box plot of Fare by Class
sns.boxplot(x='class', y='fare', data=df)
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.title('Box Plot of Fare by Class')
plt.show()
```
